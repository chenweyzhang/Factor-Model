```python
import numpy as np
import pandas as pd
import pandas_datareader as web
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

# 收集股價資料

### 可以在這邊打自己想分析的股票，^GSPC是S&P500在yahoo finance上的ticker

```python
code = ['^GSPC',  
        "AAPL",
        "MSFT",
        "DIS",
        "BRK-A",
        'AMZN',
        'JPM',
        'PG',
        'JNJ',
        'INTC',
        'UNH',
        'T',
        'VZ',
        'BAC',
        'NVDA',
        'ADBE',
       ]
```

```python
def get_price_table(code, start='2000-01-01', end='2020-12-31'):
    '''put list of stock code here
    '''
    temp=[]
    adj_close_data=[]
    for i in code:
        temp.append(web.DataReader(name='{}'.format(i), data_source='yahoo', start=start, end=end))
        print('Finding  {}'.format(i))

    for j in temp:
        adj_close_data.append( j ['Adj Close'])
        close_df = pd.concat(adj_close_data, axis=1, keys=code, join='outer')

    return close_df
```

```python
df= get_price_table(code)
```

```python
df.head()
```

# Return
將股價資料轉換成日報酬

```python
daily_ret = np.log(df/df.shift(1))
daily_ret
```

# 市場模型
將每個標的和S&P500做迴歸分析

$R_i = α + β R_M + e$

```python
alpha=[]

for i in daily_ret.columns[:0:-1]:
    regression_data= daily_ret[['^GSPC', i]].dropna()
    
    X=regression_data['^GSPC']
    X = sm.add_constant(X)
    Y=regression_data[i]
    
    model = sm.OLS(Y ,X)
    results = model.fit()
    
    print('                         from {} to {}'.format(regression_data.index[0].date() ,regression_data.index[-1].date() ))
    print('{:>60}'.format(i))
    print(results.summary().tables[1])
    print('\n')
    alpha.append([i, results.params[0]])
```

#### 依照市場模型Alpha排序

```python
alpha_rank = pd.DataFrame(alpha).sort_values(1, ascending=False)
alpha_rank = alpha_rank.set_index(0)
alpha_rank
```

```python
plot_data_index = alpha_rank.index[:5]
plot_data_index
```

```python
perform = np.exp(daily_ret.cumsum()).reset_index()
perform
```

```python
def get_info(ticker):
    regression_data= daily_ret[['^GSPC', i]].dropna()
    
    X=regression_data['^GSPC']
    X = sm.add_constant(X)
    Y=regression_data[i]
    
    model = sm.OLS(Y ,X)
    results = model.fit()

    print('{:>60}'.format(i))
    print(results.summary().tables[1])
    #---------
    sns.set_style('darkgrid')
    fig, ax = plt.subplots(figsize=(10,4))
    sns.regplot(y=Y, x=X['^GSPC'], scatter_kws={'s':14, 'alpha':0.3,'marker':'o', 'color':'orange'})
    plt.axhline(y=0, linestyle='--', color='grey')
    plt.axvline(x=0, linestyle='--', color='grey')
    ax.set_title('S&P 500 vs {}, market model in daily return'.format(i), fontsize=20)
    plt.show()
    #
    fig, ax = plt.subplots(figsize=(10,6))
    sns.lineplot('Date', '^GSPC', data=perform ,ax=ax)
    sns.lineplot('Date', ticker, data=perform, ax=ax)
    ax.set_title('S&P 500 vs {} (accumulated return)'.format(i), fontsize=20)
    ax.legend(['S&P500', i])
    plt.show()
    
```

# 對Alpha排名最高前5支股票依序跑出回歸、市場模型散佈圖以及累積報酬比較

```python
for i in plot_data_index:
    get_info(i)
```

### 三因子模型與五因子模型檢驗

```python
three_factors = pd.read_csv('F-F_Research_Data_Factors_daily.csv')
```

```python
three_factors['Date'] = pd.to_datetime(three_factors['Date'], format='%Y%m%d')
three_factors = three_factors.set_index('Date')
three_factors = three_factors/100
```

```python
three_factors
```

```python
regdata = pd.concat([daily_ret, three_factors],axis=1)
regdata = regdata.dropna()
```

```python
regdata.columns[:-4]
```

# Fama-French三因子模型超額報酬

${\displaystyle R_i - R_{f}=\alpha +\beta _{}(R_{m}-R_{f})+b_{s}\cdot {\mathit {SMB}}+b_{v}\cdot {\mathit {HML}} }$

```python
alpha_three_factor=[]

for i in regdata.columns[:-4]:
    
    X=regdata[['Mkt-RF', 'SMB', 'HML']]
    X = sm.add_constant(X)
    Y=regdata[i] - regdata['RF']
    
    model = sm.OLS(Y ,X)
    results = model.fit()
    
    print('                         from {} to {}'.format(regdata.index[0].date() ,regdata.index[-1].date() ))
    print('{:>60}'.format(i))
    print(results.summary().tables[1])
    print('\n')
    alpha_three_factor.append([i, results.params[0], results.params[1],results.params[2],results.params[3]])
```

```python
res = pd.DataFrame(alpha_three_factor)
res.columns=['Comp', "Jensen's Alpha", 'Mkt Beta', 'SMB', 'HML']
res = res.sort_values( "Jensen's Alpha", ascending=False)
res.set_index(res.Comp)
```

# Fama-French 五因子模型

${\displaystyle R_i - R_{f}=\alpha +\beta _{}(R_{m}-R_{f})+b_{s}\cdot {\mathit {SMB}}+b_{v}\cdot {\mathit {HML}}  +b_h\cdot {\mathit {RMW}}+b_{c}\cdot {\mathit {CMA}} }$

從表達式看，五因子模型比三因子模型裡面多出來了兩項：E(RMW)是高/低盈利股票投資組合的回報之差，而E(CMA)則是低/高再投資比例公司股票投資組合的回報之差。
https://zhuanlan.zhihu.com/p/21466479

```python
five_factors = pd.read_csv('F-F_Research_Data_5_Factors_2x3_daily.csv')
five_factors['Date'] = pd.to_datetime(five_factors['Date'], format='%Y%m%d')
five_factors = five_factors.set_index('Date')
five_factors = five_factors/100
five_factors
```

```python
regdata_2 = pd.concat([daily_ret, five_factors],axis=1)
regdata_2 = regdata_2.dropna()
regdata_2.columns[:-6]
```

```python
alpha_five_factor=[]

for i in regdata.columns[:-4]:
    
    X=regdata_2[['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA']]
    X = sm.add_constant(X)
    Y=regdata_2[i] - regdata_2['RF']
    
    model = sm.OLS(Y ,X)
    results = model.fit()
    
    print('                         from {} to {}'.format(regdata_2.index[0].date() ,regdata_2.index[-1].date() ))
    print('{:>60}'.format(i))
    print(results.summary().tables[1])
    print('\n')
    alpha_five_factor.append([i, results.params[0], results.params[1],results.params[2],results.params[3], results.params[4], results.params[5]])
```

```python
res = pd.DataFrame(alpha_five_factor)
res.columns=['Comp', "Jensen's Alpha", 'Mkt Beta', 'SMB', 'HML', 'RMW','CMA' ]
res = res.sort_values("Jensen's Alpha", ascending=False)
res.set_index(res.Comp)
```
